```python
# 52張鋪克牌(整體總量)
collection = 52
```

```python
# 抽出5張(預測類別)
retrieved = 5
```

```python
# 共13張方塊(實際類別)
relevant = 13
```

```python
# 真的正確(2張方塊) --- 真陽性(TruePositives)
tp = 2
print("真的正確：" + str(tp))
```

```python
# 錯的正確 --- 偽陽性(FalsePositives)
fp = retrieved - tp
print("錯的正確：" + str(fp))
```

```python
# 錯的否定 --- 偽陰性(FalseNegatives)
fn = relevant - tp
print("錯的否定：" + str(fn))
```

```python
# 真的否定 --- 真陰性(TrueNegatives)
tn = collection - (tp + fp + fn)
print("真的否定：" + str(tn))
```

```python
# 正確率
accuracy = (tp + tn) / collection
accuracy = (accuracy * 100)
print("正確率：" + str(round(accuracy,2)) + "%")
```

```python
# 錯誤率
error_rate = (fp + fn) / collection
error_rate = (error_rate * 100)
print("錯誤率：" + str(round(error_rate,2)) + "%")
```

```python
# 靈敏度/召回率(recall)
sensitivity = tp / (tp + fn)
sensitivity = (sensitivity * 100)
print("靈敏度/召回率(recall)：" + str(round(sensitivity,2)) + "%")
```

```python
# 專一性
specificity = tn / (fp + tn)
specificity = (specificity * 100)
print("專一性：" + str(round(specificity,2)) + "%")
```

```python
# 假陽性
fp_rate = fp / (fp + tn)
fp_rate = (fp_rate * 100)
print("假陽性：" + str(round(fp_rate,2)) + "%")
```

```python
# 假陰性
fn_rate = fn / (tp + fn)
fn_rate = (fn_rate * 100)
print("假陰性：" + str(round(fn_rate,2)) + "%")
```

```python
# 精確率
precision = tp / (tp + fp)
precision = (precision * 100)
print("精確率：" + str(round(precision,2)) + "%")
```
