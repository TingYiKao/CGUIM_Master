```python
import statistics as st
import numpy as np
import scipy.stats as sts

scores = [31, 24, 22, 25, 14, 25, 13, 12, 26, 23,  
          32, 34, 43, 41, 21, 23, 26, 26, 34, 42]
```

```python
# 眾數、算術、平均中位數
print('眾數：', st.mode(scores))
print('算術平均：', st.mean(scores))
print('中位數：', st.median(scores))
```

```python
#最大值、最小值與全距
print('最大值：', np.max(scores))
print('最小值：', np.min(scores))
print('全距：', (np.max(scores)-np.min(scores)))
```

```python
#樣本變異數與標準差
print('樣本變異數：',st.variance(scores))
print('樣本標準差：',st.stdev(scores))
```

```python
#樣本變異係數
print('樣本變異係數：', (st.stdev(scores) * 100 / st.mean(scores)))
```

```python
#偏度
print('偏度：', sts.skew(scores))
```

```python
#標準分數
numlist = np.array(scores)
mu = np.mean(numlist)
std = np.std(numlist)
z_score = (numlist - mu) / std
print('標準分數：', z_score)
```
