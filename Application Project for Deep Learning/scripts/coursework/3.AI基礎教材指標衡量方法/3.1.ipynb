```python
import statistics as st
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt

#資料範例
nums = [1,2,3,5,6,6,6,4,7,8]
```

```python
# 眾數
st.mode(nums)
```

```python
# 算術平均
st.mean(nums)
```

```python
# 中位數
st.median(nums)
```

```python
#百分比位數 (75%)
np.percentile(nums, 75)
```

```python
#最大值、最小值與全距
print("最大值 =",np.max(nums))
print("最小值 =",np.min(nums))
print("全距 =",np.max(nums)-np.min(nums))
```

```python
#標準差
np.std(nums)
```

```python
#樣本變異數與標準差
st.variance(nums)
st.stdev(nums)
```

```python
#母體變異數與標準差
st.pvariance(nums)
st.pstdev(nums)
```

```python
#樣本變異係數
st.stdev(nums) * 100 / st.mean(nums)
```

```python
#偏度与峰度的度量
sts.skew(nums)
```

```python
#直方圖
plt.hist(nums)
plt.show()
```

```python
#常態分布與標準常態分布
x = np.random.normal(5.0, 1.0, 100000)
y = (x-5)/1
plt.hist(x,100)
plt.hist(y,100)
plt.show()
```

```python
#標準分數
numlist = np.array(nums)
mu = np.mean(numlist)
std = np.std(numlist)
z_score = (numlist - mu) / std
print(z_score)
```
