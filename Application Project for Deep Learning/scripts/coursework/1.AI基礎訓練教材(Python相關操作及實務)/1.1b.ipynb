```python
# LSM
import numpy as np
import matplotlib.pyplot as plt

x = [45, 50, 55]
y = inFrame.iloc[:,1]  # A欄位的資料

cls = np.polyfit(x, y, deg=1)
p = np.poly1d(cls)
print("\n8.迴歸線：", end = '')
print(p)

b0=p[0]
b1=p[1]
plt.plot(x, b1 * np.array(x) + b0, color='red')
plt.scatter(x, y)
plt.show()
```
