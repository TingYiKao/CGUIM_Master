```python
# 九九乘法表

allResult = ''
for i in range(1, 10):
    lineResult = ''	
    for j in range(1, 10):
        lineResult = lineResult + str(i) + '*' + str(j) + '=' + str(i * j) + '\t'
    allResult = allResult + lineResult + '\n'

print(allResult)
```

```python
# 雙向if...else

score = 59

if score >= 60:
    print("及格！")
else:
    print("不及格！")
```

```python
# 多向if...elif...else

if score >= 90:
    print("優等")
elif score < 90 and score >= 80:
    print("甲等")
elif score < 80 and score >= 70:
    print("乙等")
elif score < 70 and score >= 60:
    print("丙等")
else:
    print("不及格")
```

```python
# NumPy統計函式

score = np.array([[80, 75, 88, 80, 78], [88, 86, 90, 95, 86], [92, 85, 92, 98, 90], [81, 88, 80, 82, 85], [75, 80, 78, 80, 70]]) 
print(score)
```

```python
# 算術平均
np.mean(score, axis = 1)
```

```python
# 加權平均
np.average(score, axis = 1, weights = [0.5, 0.2, 0.1, 0.1, 0.1])
```

```python
# 中位數
np.median(score, axis = 1)
```

```python
# 標準差
np.std(score, axis = 1)
```
