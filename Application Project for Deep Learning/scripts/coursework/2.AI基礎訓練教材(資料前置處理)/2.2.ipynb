```python
import pandas as pd
df4=pd.read_csv('df4.csv')
df4
```

```python
df5=pd.read_csv('df5.csv')
df5
```

```python
df4=df4.rename(index={0: 1, 1: 2, 2: 3})
df4
```

```python
df5=df5.rename(index={0: 2, 1: 3, 2: 4})
df5
```

```python
pd.concat([df4,df5],axis=0,join='inner')
```

```python
pd.concat([df4,df5],axis=0,join='outer')
```

```python
dfConcat=pd.concat([df4,df5],axis=0,join='outer')
dfConcat.to_csv('dfConcat.csv',index=False)
```

```python
df6=pd.read_csv('df6.csv')
df6
```

```python
df7=pd.read_csv('df7.csv')
df7
```

```python
df6=df6.rename(index={0: 1, 1: 2, 2: 3})
df6
```

```python
df7=df7.rename(index={0: 2, 1: 3, 2: 4})
df7
```

```python
# don't use the "join_axes" parameter if pandas version is older than 1.0.0 
pd.concat([df6,df7],axis=1,join_axes=[df6.index])
```

```python
# use the following function if pandas version is newer than 1.0.0 
pd.merge(df6, df7, how='left', left_index=True, right_index=True)
```
