### Example 1 (BOW)

```python
import collections
```

```python
stop_words = ['\n', 'or', 'are', 'they', 'i', 'I', 'some', 'by', '-', 'even', 'the', 'a', 'an', 'to', 'and', 'what', 'when', 'which', 'how', 'can', 'could', 'may', 'might', 'in', 'on', 'for', 'of', 'that', 'with', 'as', 'far', 'its', 'this', 'these', 'other', 'more', 'have', 'not', 'had', 'has', 'at', 'only', 'most'] 
```

```python
maxlen = 1000

word_freqs = collections.Counter()
with open('data.txt') as f:
    for line in f:
        words = line.lower().split(' ')
        if len(words) > maxlen:
            maxlen = len(words)
        for word in words:
            if word not in stop_words:
                word_freqs[word] += 1

print(word_freqs.most_common(20))
```

### Example 2 (TF-IDF)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
```

```python
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?'
]
```

```python
Vectorizer = CountVectorizer()
X = Vectorizer.fit_transform(corpus)

word = Vectorizer.get_feature_names_out()
print("Vocabulary: ", word)
```

```python
print("BOW=\n", X.toarray())
```

```python
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)
print("TF-IDF=\n", np.around(tfidf.toarray(), 4))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print("Cosine similarity: \n", cosine_similarity(tfidf[-1], tfidf[::-1], dense_output = False))
```

### Example 3 (NLTK - Text Preprocessing)

```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
text = "Today is a great day. It is even better than yesterday. And yesterday was the best day ever." + \
         " I am looking forward to tomorrow. Tomorrow will be even better than today. I am happy."
print("Text: ", text)
```

```python
nltk.sent_tokenize(text)
```

```python
nltk.word_tokenize(text)
```

```python
# Stemming
text = "My system keeps crashing his crashed yesterday, ours crashes daily"
ps = nltk.porter.PorterStemmer()
' '.join([ps.stem(word) for word in text.split()])
```

```python
# Lemmatization
text = "My system keeps crashing his crashed yesterday, ours crashes daily"
lem = nltk.WordNetLemmatizer()
' '.join([lem.lemmatize(word) for word in text.split()])
```

```python
import string

print("Punctuation: ", string.punctuation)

text = "Today is a great day. It is even better than yesterday. And yesterday was the best day ever." + \
        " And yesterday was the best day ever"

stopword_list = set(nltk.corpus.stopwords.words('english') + list(string.punctuation))

def remove_stopwords(text, is_lower_case = False):
    if is_lower_case:
        text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [token.strip() for token in tokens]
    filtered_tokens = [token for token in tokens if token not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text, filtered_tokens

filtered_text, filtered_tokens = remove_stopwords(text)
filtered_text
```

```python
# BOW model
with open('./news.txt', 'r+', encoding='utf-8') as f:
    text = f.read()

filtered_text, filtered_tokens = remove_stopwords(text, is_lower_case = True)

import collections
word_freqs = collections.Counter()

for word in filtered_tokens:
    word_freqs[word] += 1
print(word_freqs.most_common(20))
```

```python
# Regular expression
lem = nltk.WordNetLemmatizer()

def remove_stopwords_regex(text, is_lower_case = False):
    if is_lower_case:
        text = text.lower()
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+') # \w+ means any word character (equal to [a-zA-Z0-9_]) (Alphanumeric)
    tokens = tokenizer.tokenize(text)
    tokens = [lem.lemmatize(token.strip()) for token in tokens]
    filtered_tokens = [token for token in tokens if token not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text, filtered_tokens

filtered_text, filtered_tokens = remove_stopwords_regex(text, is_lower_case = True)
word_freqs = collections.Counter()

for word in filtered_tokens:
    word_freqs[word] += 1
print(word_freqs.most_common(20))
```

### Example 4 (Find Synonyms & Antonyms)

```python
import nltk
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
# Find Synonyms
synonyms = nltk.corpus.wordnet.synsets('love')
synonyms
```

```python
print(synonyms[0].definition())
print(synonyms[0].examples())
```

```python
# Find Antonyms
antonyms = []
for syn in nltk.corpus.wordnet.synsets('ugly'):
    for i in syn.lemmas():
        if i.antonyms():
            antonyms.append(i.antonyms()[0].name())
antonyms
```

```python
# POS tagging
text = "I am a human being, capable of doing terrible things"
sentence = nltk.word_tokenize(text)

for sent in sentence:
    print(nltk.pos_tag(nltk.word_tokenize(sent)))
```
